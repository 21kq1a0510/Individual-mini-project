### Bivarient & Multi-variant Analysis

```python
###1.Which species penguin has highest among other penguins
sns.countplot(penguin,x='species',hue='sex')
```

```python
###2.Species with body mass and flipper length
sns.lineplot(penguin,x='body_mass_g',y='flipper_length_mm',hue='species')
```

```python
###3.Which species is living in Torgersen island
iland=penguin[penguin['island']=='Torgersen']
sns.countplot(iland,x='species')
```

```python
###4.Who many male and female penguins in Dream island
a=penguin[penguin['island']=='Dream']
sns.countplot(a,x='sex',hue='species')
#61 female and 63 male 
```

```python
###5.Which island has highest flipper_length penguins
sns.boxplot(penguin,x='island',y='flipper_length_mm',hue="species")
```

```python
###6.How many species are living in the Biscoe island
b=penguin[penguin['island']=='Biscoe']
sns.countplot(b,x='species')
```

```python
###7.Which species have highest body_mass
sns.violinplot(penguin,x='species',y='body_mass_g')
```

```python
###8.Which species has highest culmen_length_mm 
sns.barplot(penguin,x='species',y='culmen_length_mm',estimator=np.max,hue='island')
```

```python
###9.Island wise Penguins body mass in grams
sns.kdeplot(penguin,x='body_mass_g',hue='island')
```

```python
###10.Gentoo & Chinstrap are belongs to which island
a=penguin[penguin['species'].isin(['Gentoo','Chinstrap'])]
sns.countplot(a,x='island',hue='species')
```

```python
###11.Culmen depth of species
sns.violinplot(penguin,x='species',y='culmen_depth_mm',hue='sex')
```

```python
sns.countplot(penguin,x='island',hue='species')
#gentoo is live only in biscoe
```
